#series properties and methods

```python
sales = pd.Series([150, 200, 175, 300, 250],index =['Mon','Tue','Wed','Thu','Fri'],name='Daily Sales')
print("===Series Properties ===")

#Get the actual data
print(f"Values: {sales.values}")

#Get the index Labels
print(f"Index: {sales.index}")

#Get the name of the series
print(f"Name: {sales.name}")

#Get the Dimensions
print(f"Dimensions: {sales.ndim}")

#Get the shape
print(f"Shape: {sales.shape}")

#Get the size
print(f"Size: {sales.size}")

#Get the data type
print(f"Data Type: {sales.dtype}")

```

```python
print("===Bsics Statistics ===")
print(f"Sum: {sales.sum()}")
print(f"Mean: {sales.mean()}")
print(f"Max: {sales.max()}")
print(f"Min: {sales.min()}")
print(f"Count: {sales.count()}")
print(f"standard devaiation: {sales.std()}")
print(f"Median: {sales.median()}")
```

iloc stands for index location .its a very handy tool for selecting rows and columns from dataframe by position

```python
print("===Accessing Data ===")
print(f"Monday sales: {sales['Mon']}")
print(f"Total Sales: {sales.sum()}")
print(f"Sales on Fridays: {sales['Fri']}")
print(f"first day : {sales.iloc[0]}")
print(f"last 3 days : \n{sales.tail(3)}")
```

Filtering DataFrames

```python
print("=== Filtering DataFrames ===")

#Single condition
print("Employees with salary > 50000:")
high_earners = employees[employees['Salary'] > 50000]
print(high_earners)
```

```python
#Multiple conditions (use & for ANd, | for OR)
print("IT employees with salary > 50000:")
it_high_earners = employees[(employees['Department'] == 'IT') & (employees['Salary'] > 50000)]
print(it_high_earners)
it_high_earners1 = employees[(employees['Department'] == 'IT') | (employees['Salary'] > 50000)]
it_high_earners1
```

```python
#using isin() for mltiple values
#isin() values are there checking
print("employees in IT or HR departments:")
tech_ht = employees[employees['Department'].isin(['IT', 'HR'])]
tech_ht
```

```python
# String Filtering  str.contains()
print("Employees whose names contain 'Brown' or 'johnson':")
name_filter =employees[employees['Name'].str.contains('Brown|johnson')]
print(name_filter)
```

#Data clearnig and Transformation

Handling Missing Data

```python
print('=== handling missing data ===')

#create data with missing values
messy_data = pd.DataFrame({
    'Name': ['Alice','Bob','None','Diana','Eve'],
    'Age': [25, None, 35, 28, 32],
    'Salary': [50000, 45000, None, 55000, None],
    'Department': ['IT', 'HR', 'Finance', None, 'Marketing']
})

print(messy_data)
```

```python
#detect missing values
print("Missing values info:")
print(messy_data.isnull().sum())
```

```python
#any(axis=1) means present the null values
print("Rows with any missing values:")
print(messy_data[messy_data.isnull().any(axis=1)])
```

```python
#Handling missing values

#methods 1: Drop rows with any missing values
df_dropped =messy_data.dropna()
print(df_dropped)
```

```python
#method 2: drop columns with any missing values
df_dropped_cols = messy_data.dropna(axis=1)
print(df_dropped_cols)
```

```python
#method 3:fill missing values age with mean and salry with median
df_filled = messy_data.copy()

#fill with specific values
df_filled['Name'] = df_filled['Name'].fillna('Unknown')
df_filled['Department'] = df_filled['Department'].fillna('Unassigned')

#fill with statistical measures
df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].mean())
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].median())

print(df_filled)
```

```python
#method 4 : forward fill and backward fill
time_series_data = pd.DataFrame({
    'Data': pd.date_range('2024-01-01', periods = 6),
    'Value': [100,None,None,150,None,None]
})
print("Time series with missing values:")
print(time_series_data)
```

```python
#Forward fill (use pervious value)
time_series_data['Value_ffill'] =time_series_data['Value'].fillna(method = 'ffill')

#Backward fill (use next value)
time_series_data['Value_bfill'] = time_series_data['Value'].fillna(method = 'bfill')

print(time_series_data)
```

#Data Type conversion

```python
print("=== Data Type Conversion ===")

#create sample data with wrong types
wrong_types = pd.DataFrame({
    'ID': ['1','2','3','4','5'],
    'Price': ['10.5','20.0','15.5','30.0','25.5'],
    'date': ['2024-01-01','2024-01-02','2024-01-03','2024-01-04','2024-01-05'],
    'Active': ['True','False','True','True','False']
})

print(wrong_types.dtypes)
```

```python
#convert to numeric
wrong_types['ID'] = pd.to_numeric(wrong_types['ID'])
wrong_types['Price'] = pd.to_numeric(wrong_types['Price'])
```

```python
#convert to datetime
wrong_types['date'] = pd.to_datetime(wrong_types['date'])

#convert to boolean
wrong_types['Active'] = wrong_types['Active'].map({'True': True, 'False' : False})

print(wrong_types.dtypes)
print(wrong_types)
```

```python
#Handling conversion errors
bad_data = pd.Series(['1','2','bad_value','4','5'])
print(bad_data)

#convert with error handling
converted_data = pd.to_numeric(bad_data, errors='coerce')
print(converted_data)
#
```
